```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import plotnine as ggplot
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
sns.set(style="white",color_codes=True)
df= pd.read_csv('tips.csv')
```

```python
df.head()
```

```python
df.plot(kind ="scatter", x="total_bill", y="tip")
plt.show()
```

```python
sns.jointplot(x="time",y="tip", data=df, size=5)
```

```python
sns.jointplot(x="day",y="tip", data=df, size=5)
```

```python
sns.FacetGrid(df, hue="time", size=5)\
.map(plt.scatter, "total_bill", "tip")\
.add_legend()
```

```python
sns.FacetGrid(df, hue="day", size=5)\
.map(plt.scatter, "total_bill", "tip")\
.add_legend()
```

```python
sns.boxplot(x="tip", y="time", data=df)
```

```python
sns.boxplot(x="day", y="tip", data=df)
```

```python
df.boxplot(by="day",figsize=(12,6))
```

```python
df.boxplot(by="day",figsize=(12,6))
```

```python
df.boxplot(by="day",figsize=(12,6))
```

```python
ax=sns.stripplot(x="day", y="tip", data=df , jitter=True, edgecolor="gray")
```

```python
sns.violinplot(x="day", y="tip", data= df, size=6)
```

```python
sns.FacetGrid(df, hue="day", height=6)\
.map(sns.kdeplot, "tip")\
.add_legend()
```

```python
plt.figure(figsize=(5,5))
sns.pairplot(df, hue="day", height=3)
plt.show()
```
