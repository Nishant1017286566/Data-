```python
import pandas as pd
```

```python
x=pd.read_csv("insurance.csv")
x.head()
```

```python
from plotnine import ggplot, aes, geom_line, geom_boxplot,theme_dark,geom_point,facet_grid,geom_count,labs
```

```python

(
    ggplot(x)  # What data to use
    + aes(x="region", y="bmi")  # What variable to use
    + geom_boxplot()# Geometric object to use for drawing
    + theme_dark()
)
```

```python
p = (ggplot(x)+ aes(x="charges",y="bmi",colour ="smoker") +geom_point())
```

```python
p
```

```python
z=p+ facet_grid(facets="children~")
z
```

```python
from plotnine import geom_histogram
```

```python
p=ggplot(x, aes(x="charges", colour="smoker")) + geom_histogram(bins=10)
p
```

```python
(
    ggplot(x)
    +facet_grid("sex~region")
    +aes(x="charges",color='smoker')
    +geom_histogram(bins=15)
)
```

```python
(

    ggplot(x) 

    + geom_count(mapping = aes(x='region', y='sex', color='region')) 

    + labs(title = "Number of Observations in Each Region")

)
```
